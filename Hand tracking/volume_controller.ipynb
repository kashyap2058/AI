```python
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(
    IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = interface.QueryInterface(IAudioEndpointVolume)
volume.GetMute()

# volume.GetMasterVolumeLevel()
volrange=volume.GetVolumeRange()
volume.SetMasterVolumeLevel(-20.0, None)
minvol=volrange[0]
maxvol=volrange[1]
```

```python
wcam,hcam=640,480
cap=cv2.VideoCapture(0)
cap.set(3,wcam)
cap.set(4,hcam)
ptime=0
detector=htm.handDetector(detectionCon=0.82)
vol=0
volbar=400
per=0
while True:
    sucess,img=cap.read()
    img=detector.findHands(img)
    lmlist=detector.findPosition(img,draw=False)
    if len(lmlist)!=0:
        #points for index and thumb
        thumb=lmlist[4]
        index=lmlist[8]
        xt,yt=thumb[1],thumb[2]
        xi,yi=index[1],index[2]
        cx,cy=(xi+xt)//2,(yi+yt)//2 #center points
        cv2.circle(img,(xt,yt),10,(255,0,0),cv2.FILLED)
        cv2.circle(img,(xi,yi),10,(255,0,0),cv2.FILLED)
        cv2.line(img,(xt,yt),(xi,yi),(255,0,0),3)
        cv2.circle(img,(cx,cy),10,(255,0,0),cv2.FILLED)

        #to find the length of the middle line we use hypoteous fn which will decide how much volume to control
        length=math.hypot(xi-xt,yi-yt)
        # print(length)
        #hand range 17 to 250
        # vol range -65 to 0
        vol=np.interp(length,[17,205],[minvol,maxvol])
        volbar=np.interp(length,[17,205],[400,150])
        per=np.interp(length,[17,205],[0,100])
        print(vol)
        volume.SetMasterVolumeLevel(vol, None)
        #to show that the minimum volume is reached
        if length<17:
            cv2.circle(img,(cx,cy),10,(0,250,0),cv2.FILLED)
    cv2.rectangle(img,(50,150),(85,400),(0,0,255),3)
    cv2.rectangle(img,(50,int(volbar)),(85,400),(0,0,255),cv2.FILLED)
    cv2.putText(img,f'{int(per)}%',(40,450),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),3)

    ctime=time.time()
    fps=1/(ctime-ptime)
    ptime=ctime
    cv2.putText(img,str(int(fps)),(30,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),3)
    cv2.imshow("Img",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
