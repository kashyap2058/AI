```python
import mediapipe as mp
import cv2
import numpy as np
import os
import HandTrackingFop as htf
import time
import math
```

```python
wcam,hcam=640,480
cap=cv2.VideoCapture(0)
cap.set(3,wcam)
cap.set(4,hcam)
ptime=0
prev_time=0
cur_time=0
mpHands=mp.solutions.hands
hands=mpHands.Hands(static_image_mode=False,
                   max_num_hands=2,
                   min_detection_confidence=0.85,
                   min_tracking_confidence=0.5)
mpDraw=mp.solutions.drawing_utils
fingertip=[4,8,12,16,20]
while True:
    sucess,img=cap.read()
    img,results=htf.findhand(img,0,hands,mpDraw,mpHands)
    lmlist=htf.findpos(results,img,0,draw=False)
    if len(lmlist)!=0:
        fingers=[]
        for id in range(0,5):
            if fingers[id]!=4 and (lmlist[fingertip[id]][2]<lmlist[fingertip[id]-2][2]):
                fingers.append(1)
            elif fingers[id]==4 and lmlist[4][1]<lmlist[5][1]:
                fingers.append(1)
            else:
                fingers.append(0)
        print(fingers)
    cv2.imshow("Img",img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
        
 
```
