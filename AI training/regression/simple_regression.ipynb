```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary Data.csv',usecols = ['Years of Experience','Salary'])
dataset = dataset.dropna()
dataset
```

```python
dataset['Years of Experience'] #YO garda numpy array ma hunna yeslai numpy ma lana parxa hamile always
```

```python
#converting into numpy array  
#hamilai table ko years of experience matra chaiyeko xa tesko coz YOE is independent variable where it is kept in 'X'
X = dataset.iloc[:, 0:-1].values #dataset.iloc[:,0:-1] le chai last column jun ma salary xa tyo lidaina ie takes only the value of YOE and aagadi ko : le chai savai row lyi vaneko

X


# #yo or yeslai yesari ni garna milxa
# X=dataset[['Years of Experience']].values
# X
```

```python
#Similarly hamile y ko value rakhna parxa y chai salary ho so extracting values of all rows of the last column
Y = dataset.iloc[:,1:].values
Y
```

```python
# #mathi ko iloc wala ko satta yo ni garna milxa
# Y=dataset[['Salary']].values
# Y
```

```python
#splitting datasets into train and test datasets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 10) #test_size=0.1 vaneko 10 percent data testing lai xutyako ani random_state=10 vaneko chai data lai shuffle gareko (data mining ko random sub sampling gareko)
```

```python
#just to know the total length not necessary to do
total_len=(len(X))
print('All total length =',total_len)
print('X_Train length =',len(X_train))
print('X_Test length =',len(X_test))
```

```python
#Training model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
#now the <<model>> object is prediction object
```

```python
#testing model
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]]) #hamile savai operations garne numpy array ma ho so tesma convert gareko. This is because yo case ma hamile model lai numpy ma train gareko xam 
# print(exp)

# # predict the salary
predicted_salary = model.predict(exp) #yo matra garda salary numpy ma auxa ie. [["SALARY"]] ko form ma 
predicted_salary = model.predict(exp)[0][0] #[["SALARY"]] bata "SALARY" jhikeko as salary lies in 0th index 0th element
# print(predicted_salary)
print("Your predicted_salary = ",predicted_salary)
```

```python
#to plot the regression line graph
#to plot points
plt.scatter(X_train,Y_train,color = 'red')
#yesle regression line lai banauxa as train gareko data lai model ma halera tesle deko output le regression line banauxa
plt.plot(X_train,model.predict(X_train),color = 'blue') 
plt.show()
```
