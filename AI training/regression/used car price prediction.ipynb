```python
auto=df[['symboling', 'fueltype', 'aspiration',
       'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
       'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg',
       'price']]
auto
```

```python
fuel=pd.get_dummies(auto['fueltype'],drop_first=True)
auto = pd.concat([auto,fuel], axis = 1)
auto
```

```python
a=pd.get_dummies(auto['aspiration'],drop_first=True)
auto = pd.concat([auto,a], axis = 1)
```

```python
door=pd.get_dummies(auto['doornumber'],drop_first=True)
auto = pd.concat([auto,door], axis = 1)
auto
```

```python
body=pd.get_dummies(auto['carbody'],drop_first=False)
auto = pd.concat([auto,body], axis = 1)
auto
```

```python
wheel=pd.get_dummies(auto['drivewheel'],drop_first=False)
wheel
auto = pd.concat([auto,wheel], axis = 1)
auto
```

```python
engine=pd.get_dummies(auto['enginelocation'],drop_first=True)
auto = pd.concat([auto,engine], axis = 1)
auto
```

```python
# engtype=pd.get_dummies(auto['enginetype'],drop_first=False)
# engtype
# # auto = pd.concat([auto,engtype], axis = 1)
# # auto
```

```python
cylinder=pd.get_dummies(auto['cylindernumber'],drop_first=False)
auto = pd.concat([auto,cylinder], axis = 1)
auto
```

```python
auto = auto.drop(['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','symboling'],axis = 1)
auto
```

```python
auto=auto.drop(['stroke'],axis=1)
auto
```

```python
auto=auto.drop(['carheight','compressionratio','gas','turbo','sedan','two','peakrpm','wagon','4wd','three','two'],axis=1)
```

```python
auto=auto.drop(['fuelsystem'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
df_train, df_test = train_test_split(auto, train_size = 0.85, test_size = 0.15, random_state = 10)
```

```python
df_train.shape
```

```python
df_test.shape
df_test.columns
```

```python
from sklearn.linear_model import LinearRegression
X_train = df_train[['wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginesize',
       'boreratio', 'horsepower', 'citympg', 'highwaympg',
       'convertible', 'hardtop', 'hatchback', 'fwd', 'rwd', 'rear',
       'eight', 'five', 'four', 'six', 'twelve']]
y_train = df_train['price']
# df_test_predict = X_train.iloc[0:1]
lm = LinearRegression()
model = lm.fit(X_train, y_train)

# model.predict(df_test_predict)
```

```python
# pd.get_dummies(auto['fuelsystem'],drop_first=False)
```

```python
data = df_test.iloc[14:15]
actual_price = data.pop('price')#hamilai salary predict garna parne ho tei vayera salary lai pop gareko

# print(data)

predicted_price = model.predict(data)
print("Actual price",actual_price)
print("predicted price",predicted_price)
data
# print(df_test.iloc[0:1])
```

```python
plt.scatter(y_train,model.predict(X_train),color='b')
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.show()
```

```python
#for testing datasets
X_test = df_test[['wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginesize',
       'boreratio', 'horsepower', 'citympg', 'highwaympg',
       'convertible', 'hardtop', 'hatchback', 'fwd', 'rwd', 'rear',
       'eight', 'five', 'four', 'six', 'twelve']]
y_test = df_test['price']

plt.scatter(y_test,model.predict(X_test),color='b')
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.show()
```

```python
# for testing data
predicted_pricetest = model.predict(X_test)
# predicted_salary=list(predicted_salary)
predictedtest=np.mean(predicted_pricetest)
```

```python
#only for rough estimation of the models
actualtot=np.mean(y_test)
error=(abs(predictedtest-actualtot)/actualtot)*100
print("Accuracy of the model is ",100-error)
```
